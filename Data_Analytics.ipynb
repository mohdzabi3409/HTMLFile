# **Day 01**


```python
import kagglehub
import pandas as pd
import numpy as np
```

```python
path=kagglehub.dataset_download("lava18/google-play-store-apps")
print("path to dataset files:",path)
```

```python
data=pd.read_csv(path + '/googleplaystore.csv')
print("Dataset loaded properly.")
```

```python
print("shape:",data.shape)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print("unique categ:",data['Category'].nunique())
print("sample Category:",data['Category'].unique()[:10])
```

```python
print("Duplicate rows:",data.duplicated().sum())
```

```python
data.isnull().sum()
```

```python
data['Rating'].fillna(data['Rating'].mean(),inplace=True)
```

```python
data.drop_duplicates(inplace=True)
print("Duplicate Rows removed , new shape is :",data.shape)
```

```python
data['Price']=data['Price'].astype(str).str.strip()
data['Installs']=data['Installs'].astype(str).str.strip()
data['Reviews']=data['Reviews'].astype(str).str.strip()


data['Price']=data['Price'].str.replace('$',"",regex=False)
data['Price']=pd.to_numeric(data['Price'],errors='coerce')




data['Installs']=data['Installs'].str.replace('+',"",regex=False)
data['Installs']=data['Installs'].str.replace(',',"",regex=False)
data['Installs']=pd.to_numeric(data['Installs'],errors='coerce')

data['Reviews']=pd.to_numeric(data['Reviews'],errors='coerce')

data.dropna(subset=['Price','Installs','Reviews'],inplace=True)

print("Data type cleaning done")
data[['App','Price','Installs','Reviews']].head()
```

```python
data.describe()
```

```python
data=data[data['Rating']<=5]
```

```python
data['isPaid']=data['Type'].apply(lambda x:1 if x=='Paid' else 0)
data[['App','Type','isPaid']].head()
```

```python
data.to_csv('cleaned_playstore.csv',index=False)
print("cleared Dataset Saved as cleaned_playstore.csv")
```

# Day 02

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('cleaned_playstore.csv')
```

```python
sns.set(style="whitegrid")
print("Shape:", data.shape)
```

```python
data.head()
```

```python
print(data.info())
```

```python
print("\n Missing Values:\n", data.isnull().sum())
```

```python
data.describe()
```

```python
top_categories=data['Category'].value_counts().head(10)
print("\n Top 10 Categories:\n", top_categories)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values,y=top_categories.index,palette='viridis')
plt.title('Top 10 Categories')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.show()
```

```python
avg_ratings=data.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=avg_ratings.values,y=avg_ratings.index,palette='viridis')
plt.title('Average Ratings by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

```python
app_types=data['Type'].value_counts()
print("App Types:\n", app_types)
plt.figure(figsize=(6,6))
sns.barplot(x=app_types.index,y=app_types.values,palette='coolwarm')
plt.title('App Type Distribution')
plt.xlabel('Number of Apps')
plt.ylabel('Type')
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.scatterplot(x='Price',y='Rating',data=data,color='orange')
plt.title('Price vs Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(y='Installs',x='Rating',data=data,alpha=0.6,color='green',hue='Type')
plt.xscale('log')
plt.yscale('log')
plt.title('Installs vs Rating(log scale)')
plt.xlabel('Rating')
plt.ylabel('Installs')
plt.legend(title='AppType')
plt.show()
```

```python
corr_data=data[['Rating','Reviews','Installs','Price','isPaid']].corr()
print("Correlation Matrix:\n", corr_data)

plt.figure(figsize=(8,6))
sns.heatmap(corr_data,annot=True,cmap='coolwarm',linewidths=0.5,fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
